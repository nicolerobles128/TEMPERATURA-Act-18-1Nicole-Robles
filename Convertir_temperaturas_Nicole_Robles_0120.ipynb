## 1. Fórmulas de Conversión

Para entender cómo pasar de una escala a otra, utilizamos las siguientes relaciones matemáticas:

* **De Celsius a Fahrenheit:** Multiplicamos los grados Celsius por 1.8 (o $9/5$) y sumamos 32.
    $$°F = (°C \times 1.8) + 32$$
* **De Fahrenheit a Celsius:** Restamos 32 a los grados Fahrenheit y el resultado lo dividimos entre 1.8.
    $$°C = \frac{°F - 32}{1.8}$$
---

## 2. Ejemplos Resueltos Paso a Paso

### Caso A: Convertir 25°C a Fahrenheit
1.  **Identificar datos:** Tenemos $25°C$.
2.  **Sustituir en la fórmula:** $°F = (25 \times 1.8) + 32$.
3.  **Multiplicación:** $25 \times 1.8 = 45$.
4.  **Suma final:** $45 + 32 = 77$.
**Resultado:** $25°C$ equivale a **77°F**.

### Caso B: Convertir 98.6°F a Celsius
1.  **Identificar datos:** Tenemos $98.6°F$ (temperatura corporal promedio).
2.  **Sustituir en la fórmula:** $°C = (98.6 - 32) / 1.8$.
3.  **Resta:** $98.6 - 32 = 66.6$.
4.  **División final:** $66.6 / 1.8 = 37$.
**Resultado:** $98.6°F$ equivale a **37°C**.

---

### Explicación del Código:
* **`float(input(...))`**: Permite que el usuario ingrese números decimales (necesario para temperaturas precisas).
* **`f"{variable:.2f}"`**: Es una forma de dar formato al texto para que solo muestre 2 decimales en el resultado.
* **`while True`**: Crea un bucle infinito que solo se rompe (`break`) cuando el usuario elige la opción 3.
* **`if/elif/else`**: Controla el flujo del programa según la decisión del usuario.


```python
# Función para convertir de Celsius a Fahrenheit
def convertirCF():
    celsius = float(input("Ingrese los grados Celsius: "))
    fahrenheit = (celsius * 1.8) + 32
    print(f"--- {celsius}°C equivalen a {fahrenheit:.2f}°F ---")

# Función para convertir de Fahrenheit a Celsius
def convertirFC():
    fahrenheit = float(input("Ingrese los grados Fahrenheit: "))
    celsius = (fahrenheit - 32) / 1.8
    print(f"--- {fahrenheit}°F equivalen a {celsius:.2f}°C ---")

# Función principal del Menú
def menu():
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir grados Centígrados a Fahrenheit")
        print("2. Convertir grados Fahrenheit a grados Centígrados")
        print("3. Salir")

        opcion = input("Seleccione una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa. ¡Hasta luego!")
            print("!Nicole Robles NL = 46¡")
            break
        else:
            print("Opción no válida, por favor intente de nuevo.")

# Ejecución del programa
if __name__ == "__main__":
    menu()
```
